```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np

# plot via pandas
s1 = pd.Series(range(10))
print(s1)

print()
s2 = s1**2
print(s2)

plt.plot(s1,c='r',alpha=0.5, linewidth=4) # alpha => transparency float : [0-1]
plt.show()

plt.plot(s2)
plt.show()
```

### subplotting and labeling things

```python
# plt.subplot(x,y, idx)
# <no of row>, <no. of col> , < position in grid>

# plt.subplot (2,3,1)
# 1 - -
# - - 6
```

```python
s1 = pd.Series(range(10))
print(s1)

print()
s2 = s1**2
print(s2)
```

```python
#  - -
plt.subplot(1,2,1)
plt.plot(s1) # alpha => transparency float : [0-1]

plt.subplot(1,2,2)
plt.plot(s2)
plt.show()
```

```python
plt.figure(figsize=(16,6))
# 1  2
# 3  4
plt.subplot(2,2,1)
plt.plot(s1, c= 'r') # alpha => transparency float : [0-1]

plt.subplot(2,2,2)
plt.plot(s2, c='g')

plt.subplot(2,2,4)
plt.plot(s2//2)
plt.plot(s1)

plt.subplot(2,2,3)
plt.plot(s2+s1)
plt.show()
```

```python
# labeling
plt.figure(figsize=(16,6))
plt.plot(s1, c= 'r') 
plt.xlabel("x - axis")
plt.ylabel("y - axis")
plt.title("A graph")
```

```python
# labeling
plt.figure(figsize=(16,6))
s1.plot(kind='line') 
plt.xlabel("x - axis")
plt.ylabel("y - axis")
plt.title("A graph")
plt.show()
```

```python
# labeling
plt.figure(figsize=(16,6))
s1.plot(kind='line') 
plt.xlabel("x - axis")
plt.ylabel("y - axis")
plt.title("A graph")
plt.yticks(range(10))
plt.show()
```
