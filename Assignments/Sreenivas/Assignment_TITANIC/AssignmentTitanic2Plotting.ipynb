```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(r"c:\Users\srr07\Desktop\GitPython\dsc_weekday_2020_09_24\Assignments\Sreenivas\Assignment_TITANIC\titanic_Dataset.csv")
```

```python
# plot histogram of ages of people
```

```python
plt.hist(df.age, width = 5)
```

```python
#plot no of people in each pclass on bar chart
group=df.groupby(['pclass'])
print(group.size())
a=group.size()
a.plot(kind='bar')
```

```python
# plot a bar-chart depicting the ratio of males to females
# plot a horizontal bar-chart depicting the ratio of males to females
group=df.groupby(['gender','survived'])
print(group.size())
a=group.size()
a.plot(kind='bar')
```

```python
# plot a bar-chart depicting the ratio of male-survived, female-survived, male-dead, female-dead.
group=df.groupby(['gender','survived'])
print(group.size().unstack())
a=group.size()
a.plot(kind='bar',stacked=True)
```

```python
# plot a stacked bar chart of number of male-survived,dead and female-survived, dead.
#plot a stacked bar chart of number of male-survived,dead and female-survived, dead.
group=df.groupby(['gender','survived'])
print(group.size())
a = group.size().unstack()
print(a)
a.plot(kind='bar',stacked=True,alpha=0.4) 
```

```python
# plot stacked bar of survived with different pclass in same bar
group=df.groupby(['pclass'])
print(group.size())
a = group.size()
print(a)
a.plot(kind='bar') 
```
