```python
import pandas as pd
```

```python
def conver_na(cell):
    if cell == '..':
        return None
    return cell
xl= pd.read_excel(r'C:\Users\Akilan\Documents\Python Scripts\Assignment_IMMIGRANTS\Australia.xlsx',1,skip_rows=19,header=20,converters={
       2009 : conver_na,
       2010 : conver_na,
       2011 : conver_na,
       2012 : conver_na,
       2013 : conver_na,
    
})
print(xl)
```

```python
print(xl.index)
print(xl.columns)
xl.describe(include='all')
```

```python
print(xl.shape)
print(xl.dtypes)
print(xl.size)
```

```python
xl.OdName.values
```

```python
xl=xl[xl.Type == 'Immigrants']
print(xl)
```

```python
xl=xl.dropna(thresh=xl.shape[0]*0.5,axis=1)
```

```python
xl.isna().sum()
```

```python
#Drop Unnessary Columns: Type, Coverage, Area, Reg, Dev 
xl.drop(columns=['Type','Coverage','AREA','REG','DEV'],inplace=True)
```

```python
print(xl.head())
```

```python
#Rename columns :OdName -> Country; RegName -> Region
xl.rename(columns={"OdName":"Country","RegName": "Region"},inplace=True)
print(xl.head())
xl.columns.map
```

```python
#- Convert Column Names with Int type to Strings to avoid index errors with string indexing "1980, 1981 etc." => " '1980', '1981' ... " (Use rename)
xl.rename(columns= lambda x:str(x),inplace=True)
print(xl.head())
```

```python
xl.columns.map
```

```python
xl1 = xl
xl1['Total'] = xl.iloc[:,4::].sum(axis=1)
print(xl1.head())
```

```python
xl1.sort_values(by="Total",ascending=False,inplace=True)
print(xl1.head())
```

```python
xl1 = xl1[xl1.Region !='World']
```

```python
xl1.to_excel(r'C:\Users\Akilan\Documents\Python Scripts\Assignment_IMMIGRANTS\AustraliaFiltered.xlsx','Head')
```
