```python
import pandas as pd
import matplotlib.pyplot as plt
df1= pd.read_excel(r'C:\Users\Akilan\Documents\Python Scripts\Assignment_IMMIGRANTS\AustraliaFiltered.xlsx',0)
```

```python
df2=df1.head(5)
df2= df2.loc[:,["Country","Total"]]
df2.plot(x="Country",y="Total",kind='bar')
#plt.bar("Country","Total")
plt.show()
```

```python
grp=df1.groupby(["AreaName","Country"])
r=grp.Total.sum()
r1=r.unstack(level="Country")
r1
```

```python
df2=df1.T
df3=df2.iloc[5:len(df2.index)-1,:]
df3.fillna(0)
df3['Total']=df3.iloc[0:,:].sum(axis=1)
plt.figure(figsize=(20,10))
plt.plot(df3.index,df3.Total)
plt.show()
```

```python
df4=df1[df1.Country=='India']
df2=df4.T
df3=df2.iloc[5:len(df2.index)-1,:]
df3.fillna(0)
df3['Total']=df3.iloc[0:,:].sum(axis=1)
plt.figure(figsize=(20,10))
plt.plot(df3.index,df3.Total)
plt.show()
```
