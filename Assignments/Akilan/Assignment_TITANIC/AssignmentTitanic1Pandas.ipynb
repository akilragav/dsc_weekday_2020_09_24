```python
import pandas as pd
import numpy as np
#1. read and Describe
df1= pd.read_csv('titanic_filtered.csv')
print(df1.head())
df1.describe(include='all')
```

```python
#2. find a columns with missing data
df1.isna().any()
```

```python
#3. Shape of dataset
print(df1.shape)
```

```python
#4. drop a column with more than 25% na
print(df1.dropna(thresh=df1.shape[0]*0.75,axis=1).head())
```

```python
#5. Check datatype of all column
print(df1.dtypes)
```

```python
#6. Convert price to float
df1.fare=df1.fare.str.replace('$','').astype('float')
print(df1,df1.dtypes)
```

```python
#7 find a colum still having missing Na and Count
print(df1.isna().sum(axis=0) )
```

```python
#8 fill na with mean for fare and age column column
print(df1)
df1.fillna({"fare": df1.fare.mean() , "age": df1.fare.mean()}, inplace=True)
print(df1.isna().sum(axis=0) )
```

```python
#drop na emarked values
df1=df1[pd.notnull(df1['embarked'])]
df1.count()
```

```python
df1.to_csv( 'titanic_filtered.csv',index=False)
```

```python
df1.survived.replace({0:'D', 1:'A'},inplace=True)
```

```python
df1.survived.value_counts()
```

```python
grp=df1.groupby(["survived","gender"])
grp.size()
```

```python
grp1=df1.groupby("pclass")
grp1.count()
```

```python
grp['age'].agg(np.max)
```

```python
grp2=df1[df1.survived=='A'].groupby('gender')
print(grp2.age.max())
```

```python
grp3=df1.groupby('gender')
print(grp3.age.mean())
```
