```python
import pandas as pd


# Read data and describe it

a = pd.read_csv("titanic_dataset.csv")
a.describe()
```

```python
a.describe(include='all')
```

```python
print(a.columns)
print(len(a.columns))
print(a.shape) 
```

```python
# Find columns with missingdata

a.isna().any()
```

```python
# print shape of dataset

a.shape
```

```python
# drop columns with more than 25% missing data

a.dropna(thresh = a.shape[0]*0.75, axis=1, inplace=True)
a.head()
```

```python
# drop columns having independent values(which do not affect the survival rate)

a.drop(columns=["Unnamed: 0"], inplace=True)
a.columns
```

```python
# check data types of all columns

a.dtypes
```

```python
# convert price to numeric

a.fare.replace(r"\$","", regex=True).astype('float')       # using regular expression (recommended)
a.head()
```

```python
# find columns still having missing/na values and also count of missing data

a.isna().any()
```

```python
# drop na values for embarked column.

a.dropna(inplace=True)
a.shape
```

```python
a.count()
```

```python
# dump the dataframe to a csv file 'titanic_filtered.csv'.
a.to_csv(r"C:\Users\Ketan Jamode\Desktop\dsc_weekday_2020_09_24\Assignments\Ketan\Titanic\titanic_filtered.csv", index = False )
```

```python
# for surviced column replace 0 with D and 1 with A

a.survived.replace({0:"D", 1 :"A"})
```

```python
# find the frequency of different values in survived column

a.survived.value_counts()
```

```python
# group by gender and survived and see the counts in each category

grp = a.groupby(["gender", "survived"])
grp.groups
```

```python
# find different pclass and no of people in each class

grp = a.groupby("pclass")
grp.pclass.count()
```

```python
# find top 5 people with highest values of age. Count no of male and females in the top 5

grp = a.groupby("age")
grp.max()
```

```python
# find max age male and female who survived

grp = a.groupby(["gender"])
grp.age.max()
```

```python
# get average age by gender
grp = a.groupby("gender")
grp.age.mean()
```

```python
# get average age by people survived vs not-survived
grp = a.groupby("survived")
grp.age.mean()
```
