```python
#Read data and describe it
import pandas  as pd
df =  pd.read_csv(r'titanic_dataset.csv')
df.describe(include='all')
```

```python
#Find columns with missingdata
print(df.isna().any())
print()
# count no  of nan values  in each column
print(df.isna().sum())
print()
print(df.count())
print()
# percentage  missing data
print(df.count()/df.shape[0]*100)
print()
# percentage  missing data
print(df.isna().sum()/df.shape[0] * 100)
```

```python
#Print shape of dataset
df.shape
```

```python
#drop columns with more than 25% missing data
print(df.shape[0]*0.75)
r = df.dropna(thresh=981, axis=1)
r.head()
```

```python
df.head()
```

```python
df.dropna(thresh=0.75*(len(df.values)),axis=1, inplace=True)
```

```python
#drop columns having independent values(which do not affect the survival rate).
df.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
#check data types of all columns
df.dtypes
```

```python
#convert price to numeric
# slicing
df.fare.str[1:].astype('float')

# str.replace
df.fare.str.replace("$", "").astype('float')

# replace
df.fare = df.fare.replace(r"\$", "", regex=True).astype('float')
```

```python
#find columns still having missing/na values and also count of missing data
#Find columns with missingdata
print(df.isna().any())
print()
# count no  of nan values  in each column
print(df.isna().sum())
```

```python
#fill na with mean for fare and age column column.
df.fillna({'fare':df.fare.mean(), 'age': df.age.mean()}, inplace=True)
print(df.isna().sum())
```

```python
#drop na values for embarked column.
df.dropna(inplace=True)
df.count()
```

```python
#dump the dataframe to a csv file 'titanic_filtered.csv'.
df.to_csv (r'titanic_filtered.csv')
```

```python
#for surviced column replace 0 with D and 1 with A
f = lambda x: "D" if '0' in x else "A"
df['survived']=df.survived.astype('str').apply(f) # python: map 
```

```python
#find the frequency of different values in survived column
df.survived.value_counts()
```

```python
#group by gender and survived and see the counts in each category
grp1 = df.groupby("survived")
grp2 = df.groupby("gender")
```

```python
grp2.groups.keys()
```

```python
grp1.count()
```

```python
grp2.count()
```

```python
#find different pclass and no of people in each class
grp = df.groupby("pclass")
grp.groups.keys()
```

```python
grp.count()
```

```python
#find top 5 people with highest values of age. Count no of male and females in the top 5
df.nlargest(5, ['age'])
```

```python
df.nlargest(5, ['age']).gender.value_counts()
```

```python
#find max age male and female who survived
grp = df.groupby(["gender","survived"])
grp.get_group( ('male',   'A') ).nlargest(5, ['age'])
```

```python
grp = df.groupby(["gender","survived"])
grp.get_group( ('female',   'A') ).nlargest(5, ['age'])
```

```python
#get average age by gender
grp = df.groupby(["gender"])
grp.age.mean()
```

```python
#get average age by people survived vs not-survived
grp = df.groupby(["survived"])
grp.age.mean()
```
