```python
#7 Delete all Data where the Type = "Emigrants" or Type is not "Immigrants" (Row Deletion).

rdel1=ds1[ds1.Type=='Emigrants'].index
ds1.drop(rdel1, inplace=True)
ds1
```

```python
#8 Find columns for which more than 50% data is missing and drop them (Column Deletion)
r=ds1.isna().sum()/ds1.shape[0]*100
print(r)
ds1.dropna(thresh=ds1.shape[0]*0.75,axis=1, inplace=True)
ds1.head()
```

```python
ds1.shape
```

```python
#Drop Unnessary Columns: Type, Coverage, Area, Reg, Dev (Column deletion) Your columns should now look like "OdName,AreaName,RegName,DevName,1980 ... 2008"
ds1.drop(['Type', 'Coverage', 'AREA', 'REG', 'DEV'], axis=1, inplace=True)
ds1
```

```python
#10 Rename columns :OdName -> Country; RegName -> Region
ds1.rename(columns={"OdName": "Country", "RegName": "Region"}, inplace=True)
ds1
```

```python
#12. Find Total number of immigrants for all years, and create a new column Total in data Frame
z=ds1.columns[4:38]
ds1['Total']=ds1[z].sum(axis=1)
ds1
```

```python
#11. Convert Column Names with Int type to Strings to avoid index errors with string indexing "1980, 1981 etc." => " '1980', '1981' ... " (Use rename)
ds1.rename(columns={1980:"1980"},inplace=True)
ds1.dtypes


```
