```python
#Delete all Data where the Type = "Emigrants" or Type is not "Immigrants" (Row Deletion)
imm.drop(imm.index[imm.Type=="Emigrants"] | imm.index[imm.Type !="Immigrants"],inplace=True)
imm
```

```python
#Find columns for which more than 50% data is missing and drop them
imm.replace('..',np.nan,inplace = True)
print(imm)
imm.isna().sum()
imm.shape[0]
imm.dropna(thresh = 0.5*(imm.shape[0]),axis=1,inplace = True)
imm
```

```python
#Drop Unnessary Columns: Type, Coverage, Area, Reg, Dev (Column deletion)
imm.drop(columns=["Type","Coverage","AREA","REG","DEV"],inplace = True)
imm
```
